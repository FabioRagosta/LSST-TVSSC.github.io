```python
# coding: utf-8
#just prints the emails of members of a group to stdout,
#both primary and secondary members
# run as
# $python extractemails_nogui.py "Tidal Disruption Events"

from __future__ import print_function
'__author__' == 'Federica Bianco, NYU - GitHub: fedhere'
import sys
import pandas as pd
import numpy as np
tvsfile = '1Q8YPGxh-Ebs02ykfGypj7oThQrgqKzrU6qyIuPT3bhY'
primary = "supernova".lower()
secondary = "TIDAL DISRUPTION EVENTS".lower()

```

```python

sbuglist = ['']

TVSMembers = pd.read_csv('https://docs.google.com/spreadsheets/d/' +
                             tvsfile +
                             '/export?gid=0&format=csv')
TVSMembers.rename({"Unnamed: 0":"last name"}, axis=1, inplace=True)
TVSMembers = TVSMembers.loc[~TVSMembers["signed CoC"].isna()]
TVSMembers.head()
```

```python
subgroups = list(set(np.concatenate([TVSMembers.primary.unique(), 
                                     TVSMembers["secondary"].unique(), 
            TVSMembers["secondary.1"].unique(), 
            TVSMembers["secondary.2"].unique()])))[1:]
subgroups
```

```python
import glob
import ipywidgets as widgets
from IPython.display import display
from IPython.html.widgets import interactive

def primarylist(x):
  print("primary selected:", x)
  # PRIMARY
  emails = TVSMembers[TVSMembers.primary == x]['email'].values
  print(emails)
  return x

w = widgets.Select(options=subgroups)
interactive(primarylist, x=w)
```

```python

def secondarylist(x):
  print("secondary selected:", x)
  
  emails = TVSMembers[TVSMembers.secondary == x]['email'].values
  print(emails)
  return x

w = widgets.Select(options=subgroups)
interactive(secondarylist, x=w)
```
